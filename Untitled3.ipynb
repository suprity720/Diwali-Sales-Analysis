```python
# Create a histogram of the "Age vs Attrition" feature
plt.figure(figsize = (9,4))

sns.histplot(data = df ,x = 'Age', color = 'Orange',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Age vs Attrition")
plt.xlabel('Age')
plt.ylabel('Attrition')
plt.show()
```

## mostly employees who is having a age group between 27 to 43 who left the job.

```python
## Create a histogram of the "YearsAtCompany vs Attrition" feature

plt.figure(figsize = (9,4))

sns.histplot(data = df ,x = 'YearsAtCompany', color = 'Orange',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("YearsAtCompany vs Attrition")
plt.xlabel('YearsAtCompany')
plt.ylabel('Attrition')
plt.show()
```

Mostly, employees who have worked from 1 to 7 years have left the company (approximately 900 employees). Subsequently, employees who have worked from 8 to 16 years have also left the company (approximately 300 employees).

```python
## Create a histogram of the "MonthlyIncome vs Attrition" feature
plt.figure(figsize = (9,4))

sns.histplot(data = df ,x = 'MonthlyIncome', color = 'Orange',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("MonthlyIncome vs Attrition")
plt.xlabel('MonthlyIncome')
plt.ylabel('Attrition')
plt.show()




```

Employees earning a salary of range 1500 to 3000 have left the job, approximately 900 in number. As the salary increases, the attrition rate decreases.







```python
## what is the attrition rate (% of people who have left)
```

```python
df['Attrition'].value_counts()
```

```python
# Count the occurrences of 'Yes' and 'No' in the 'Attrition' column
attrition_counts = df['Attrition'].value_counts()

# Calculate the attrition rate
attrition_rate = attrition_counts['Yes'] / df['Attrition'].count()

print("Attrition Rate:", attrition_rate)
```

Attrition rate is 0.16122448979591836

**Conclusion:**
The data indicates that younger employees within the age group of 27 to 43, especially those with 1 to 7 years of tenure and earning salaries between 1500 to 3000, are more likely to leave the company. This suggests that factors such as career advancement opportunities, competitive compensation, and work-life balance may be significant drivers of attrition within this demographic. Therefore, the company should focus on implementing strategies to retain and engage these employees, such as offering opportunities for career growth, providing competitive salaries and benefits, and promoting a positive work culture that prioritizes work-life balance.
